```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from scipy.interpolate import Akima1DInterpolator
import numpy as np

# Data
data = {
    "time": ["5:00 AM", "8:00 AM", "1:00 PM", "5:00 PM", "9:00 PM"],
    "mon": [60.97, 72.94, 68.92, 82.93, 66.93],
    "wed": [61.98, 73.95, 68.96, 73.94, 67.95],
    "fri": [61.97, 70.97, 69.94, 74.92, 76.97],
    "sat": [65.98, 68.94, 68.92, 74.94, 68.95],
    "sun": [67.90, 68.99, 67.99, 73.94, 77.99],
}

df = pd.DataFrame(data)

# Convert time to numerical scale for interpolation
time_numeric = np.arange(len(df["time"]))

# Interpolate data for smooth curves
def interpolate_line(x, y):
    x_new = np.linspace(min(x), max(x), 500)
    f_akima = Akima1DInterpolator(x, y)
    y_new = f_akima(x_new)
    return x_new, y_new

# Plot
plt.figure(figsize=(12, 6))
for day in ["mon", "wed", "fri", "sat", "sun"]:
    x_new, y_new = interpolate_line(time_numeric, df[day])
    plt.plot(x_new, y_new, label=day, alpha=0.8)
    plt.scatter(time_numeric, df[day])

    # plt.scatter(time_numeric, df[day], label=f"{day} points")

# Customize the plot
plt.title("Uber Cost vs. Time of Day Across Days", fontsize=16)
plt.xlabel("Time of Day", fontsize=12)
plt.ylabel("Money ($)", fontsize=12)
plt.xticks(time_numeric, df["time"], rotation=45)
plt.legend(title="Day", fontsize=10)
plt.grid(alpha=0.3)
# plt.tight_layout()

# Show the plot
plt.show()
```
